```python
# Cell 1: Imports
import sys
sys.path.append('..')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs, load_iris
from sklearn.preprocessing import StandardScaler

from src.models.kmeans import KMeansModel
from src.models.hierarchical_clustering import HierarchicalClusteringModel
from src.models.pca import PCAModel

plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("✅ All imports successful!")
```

```python
# Cell 2: Load and Prepare Data
print("="*60)
print("📊 LOADING DATA FOR UNSUPERVISED LEARNING")
print("="*60)

# Synthetic data for clustering
X_blobs, _ = make_blobs(n_samples=300, centers=3, random_state=42)

# Iris dataset for PCA
X_iris, y_iris = load_iris(return_X_y=True)

# Scale data
scaler = StandardScaler()
X_blobs_scaled = scaler.fit_transform(X_blobs)
X_iris_scaled = scaler.fit_transform(X_iris)

print(f"Blobs dataset shape: {X_blobs_scaled.shape}")
print(f"Iris dataset shape: {X_iris_scaled.shape}")
```

```python
# Cell 3: K-Means Clustering
print("\n" + "="*60)
print("📊 K-MEANS CLUSTERING")
print("="*60)

kmeans = KMeansModel(n_clusters=3)
kmeans.train(X_blobs_scaled)
results = kmeans.evaluate(X_blobs_scaled)
kmeans.find_optimal_k(X_blobs_scaled, max_k=8)
kmeans.plot_clusters(X_blobs_scaled, title="K-Means Clustering Results")
```

```python
# Cell 4: Hierarchical Clustering
print("\n" + "="*60)
print("📊 HIERARCHICAL CLUSTERING")
print("="*60)

hierarchical = HierarchicalClusteringModel(n_clusters=3, linkage='ward')
hierarchical.train(X_blobs_scaled)
results = hierarchical.evaluate(X_blobs_scaled)
hierarchical.find_optimal_clusters(X_blobs_scaled, max_clusters=8)
hierarchical.plot_dendrogram(X_blobs_scaled)
hierarchical.plot_clusters(X_blobs_scaled, title="Hierarchical Clustering Results")
```

```python
# Cell 5: PCA for Dimensionality Reduction
print("\n" + "="*60)
print("📊 PCA ANALYSIS")
print("="*60)

pca = PCAModel(n_components=2)
X_iris_reduced = pca.fit_transform(X_iris_scaled)
results = pca.evaluate(X_iris_scaled)
pca.plot_explained_variance()
pca.plot_2d_projection(X_iris_scaled, y=y_iris, title="Iris Dataset - PCA 2D Projection")
```

```python
# Cell 6: Compare Clustering Methods
print("\n" + "="*60)
print("📊 COMPARING CLUSTERING METHODS")
print("="*60)

# K-Means
kmeans_scores = []
for k in range(2, 8):
    km = KMeansModel(n_clusters=k)
    km.train(X_blobs_scaled)
    results = km.evaluate(X_blobs_scaled)
    kmeans_scores.append(results['silhouette_score'])

# Hierarchical
hierarchical_scores = []
for k in range(2, 8):
    hc = HierarchicalClusteringModel(n_clusters=k, linkage='ward')
    hc.train(X_blobs_scaled)
    results = hc.evaluate(X_blobs_scaled)
    hierarchical_scores.append(results['silhouette_score'])

# Plot comparison
plt.figure(figsize=(10, 6))
k_range = range(2, 8)
plt.plot(k_range, kmeans_scores, 'bo-', label='K-Means', linewidth=2)
plt.plot(k_range, hierarchical_scores, 'ro-', label='Hierarchical', linewidth=2)
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Clustering Methods Comparison')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\n✅ K-Means Silhouette Scores:", kmeans_scores)
print("✅ Hierarchical Silhouette Scores:", hierarchical_scores)
```
